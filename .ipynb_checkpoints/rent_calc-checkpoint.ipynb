```python
from IPython.display import clear_output
class RentCalc():
    rent_amount = 0
    investment = 0
    return_investment = 0

    #anything that needs to get passed between functions or from the logic function to this set
    def __init__(self, total_monthly_income, total_monthly_exp, monthly_cash_flow,):
        self.total_monthly_income = total_monthly_income
        self.total_monthly_exp = total_monthly_exp
        self.monthly_cash_flow = monthly_cash_flow
        
    def calcIncome(self): 
        
        self.rent_amount = self.formatting(input('How much is rent per month? '))
             
        add_income = input('Do you have any additional income from this property? Yes or no: ')
        add_income_amount = 0
        self.total_monthly_income = self.rent_amount

        while add_income.lower() in ['yes','yeah','yea','y']:
            add_income_item = input('What is the source of the income? ')
            add_income_amount = self.formatting(input(f'How much income does {add_income_item} generate each month? '))
            self.total_monthly_income += add_income_amount
            add_income = input('Are there any other sources of income? ')
        clear_output()
        
        return self.total_monthly_income
    
    def calcExp(self): 
        taxes = self.formatting(input('What are the monthly taxes for the property? '))
        insurance = self.formatting(input('What is the monthly insurace cost for the property? '))
        mort = self.formatting(input('What is the monthly mortgage amount? '))
        
        # Repairs
        repair_cost = self.formatting(input('How much do you save monthly for regular repairs? '))
        cap_costs = self.formatting(input('How much do you save monthly for major repairs? '))

        # Property Management
        print('Is there a property manager managing this property?')
        prop_check = input('Yes or no: ').lower()
        if prop_check in ['yes','yeah','yea','y']:
            print('A typical amount for a property manager is 10% of the rent. For this property, that would be', self.rent_amount * .1)
            prop_amount = self.formatting(input('How much is the monthly cost of management? '))
            
        # Vacancy
        # user to accept that amount or enter a different amount
        print('How much do you save monthly to cover potential vacancies?', 
              '\nA typical amount is 5% of the monthly rent, which would be',self.rent_amount * .05)
        print('Do you accept this amount, or would you like to enter your own vacancy amount?')
        
        vacancy_check = input('Enter accept, own to add your own, or no to skip: ').lower()
        if vacancy_check in ['accept','yes','y']:
            vacancy_cost = self.rent_amount * .05
        elif vacancy_check == 'own':
            vacancy_cost = self.formatting(input('Enter your monthly vacancy cost: '))
        elif vacancy_check in ['no','none','n','skip']:
            vacancy_cost = 0
            

        # zeroing out optional expenses
        hoa_cost = 0
        utility_cost = 0

        # getting HOA costs
        hoa_check = input('Are there any HOA costs you pay? Yes or no: ')
        if hoa_check.lower() == 'yes':
            hoa_cost = self.formatting(input('What are the monthly HOA costs? '))

        # getting utility costs
        utility_check = input('Who covers the cost of utilities, you or the tenant? Enter self or tenant: ')
        if utility_check.lower() == 'self':
            utility_cost = self.formatting(input('What is the average monthly cost of all utilities electric, water, sewer, gas, trash, etc)? '))

        # total monthly exp pre user added exps
        self.total_monthly_exp = taxes + insurance + mort + repair_cost + cap_costs + prop_amount + vacancy_cost + hoa_cost + utility_cost  

        # gets as many extra exps as needed
        add_exp_check = input('Are there additional monthly costs? For example, lawn care, snow removal, etc? Enter yes or no ')
        while add_exp_check.lower() == 'yes':
            add_exp_item = input('What is the name of the additional expense? ')
            add_exp_cost = self.formatting(input(f'What is the monthly cost of {add_exp_item}? '))
            
            self.total_monthly_exp = self.total_monthly_exp + add_exp_cost
            add_exp_check = input('Are there additional expenses? Enter yes or no: ')
        clear_output()
        return self.total_monthly_exp
        
    def cashFlow(self):
        self.monthly_cash_flow = self.total_monthly_income - self.total_monthly_exp
        return self.monthly_cash_flow
    
    def totalInvestment(self):
        down_payment = self.formatting(input('How much was the down payment on the property? '))
        closing = self.formatting(input('How much were the closing costs? '))
        rehab = self.formatting(input('How much did you invest into rehab projects prior to renting the property? '))
        
        self.investment = down_payment + closing + rehab       
        
        add_costs = input('Were there any additional costs? Enter yes or no: ').lower()
        while add_costs == 'yes':
            add_cost_item = input('What was the name of additional cost? ')
            add_cost_exp = self.formatting(input(f'What was the cost of {add_cost_item}? '))
            
            self.investment += add_cost_exp
            add_costs = input('Were there any additional costs? ')
            
        return self.investment
    
    def cashReturn(self):
        yearly_cash = self.monthly_cash_flow * 12
        cash_yearly_cash = '${:,}'.format(yearly_cash)
        cash_investment = '${:,}'.format(self.investment)
        
        print(cash_yearly_cash,'is your Annual Cash Flow')
        print(cash_investment,'is your Total Investment cost')
        self.return_investment = yearly_cash / self.investment
        return self.return_investment
        
    def formatting(self, user_input):
        if user_input.isalpha():
            message = 'Please only enter numbers in this field'
            return message
        if ',' in user_input:
            user_input = user_input.replace(',','')
        if user_input.startswith('$'):
            user_input = user_input.replace('$','')
        user_input = float(user_input)
        return user_input   
```

```python
rent = RentCalc(0,0,0)
def calcRent():
    print('Enter your income information to calculate your rental income percentage')
    RentCalc.calcIncome(rent)
    cash_income = "${:,}".format(rent.total_monthly_income)
    print('Your monthly income is',cash_income)

    print('\nThis process will calculate your monthly expenses for the property. If none, answer 0: ')
    RentCalc.calcExp(rent)
    cash_exp = '${:,}'.format(rent.total_monthly_exp)
    print('Your monthly expenses are',cash_exp)
    
    RentCalc.cashFlow(rent)
    cash_flow = '${:,}'.format(rent.monthly_cash_flow)
    print('Your total monthly cash flow is', cash_flow)
    
    RentCalc.totalInvestment(rent)
    cash_investment = '${:,}'.format(rent.investment)
    print('Your total investment in the property is', cash_investment)
    
    RentCalc.cashReturn(rent)
    return_dec = rent.return_investment
    return_statement = "Your cash on cash ROI is: {:.0%}"
    print(return_statement.format(return_dec))
    
calcRent()
```
